```python
#Code No.10
#Code for displaying the "Name", "Nationality" and "Age" of five youngest and five eldest players.

import pandas as pd
import numpy as np
a = pd.read_csv('players.csv')
a['Age'].value_counts()
col = ['ID', 'Name', 'Age', 'Nationality']
c = pd.DataFrame(a, columns = col)
d_top5_sorted = c.groupby(['Age']).apply(lambda x:x.sort_values(by = 'Age', ascending = True, axis =0).head(5))
print('\n---------Five youngest and Five eldest players-----------\n')
d_top5_sorted
```

```python
#Objective No.11:Visualize top-10 football playing nation using bar chart.
```

```python
#Code No.16
#Code for visualizing the top-10 foorball playing nations using bar chart

import pandas as pd
import matplotlib.pyplot as plt
a = pd.read_csv('players.csv')
print("\n--------Visualizing the top-10 football playing nations-----------\n")
a['Preferred Foot'].value_counts().index
col = ['Nationality']
c = pd.DataFrame(a, columns = col)
d_top10_sorted = c.groupby(['Nationality']).apply(lambda x:x.sort_values(by = 'Nationality', ascending = False, axis = 0).head(10))

plt.hist(d_top10_sorted)
plt.xlabel('Nations')
plt.ylabel('No. of Players')
plt.plot()
plt.show()
```

```python
#Objective No.12:Visualization of top-n teams with best players (where n=5) with their percentage contribution using pie chart.
```

```python
#Code No.17
#Code for visualization of top-n teams with best players (where n = 5) with their percentage contribution using pie chart.

import pandas as pd
import numpy as np
a = pd.read_csv('players.csv')
print("\n---------Visualization of top-n teams with best players (where n = 5) with their percentage contribution------------\n")
import matplotlib.pyplot as plt
labels = a['Overall'].value_counts(5)
size = a['Overall'].value_counts(5)
colors=['cyan','red','green','blue']
plt.pie(size, labels = labels, colors = colors, shadow = True, autopct = '%1.1f%%', startangle = 90)
plt.title('Best Players', fontsize = 26)
plt.show()
```

```python
#Code No.18
#Code for visualizing the co-relation between overall performance and acceleration of players.

import pandas as pd
a=pd.read_csv('players.csv')
print("\n-----------Visualizing the co-relation between overall performance and acceleration of players-----------\n")
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']=(9,5)
matplotlib.rcParams['figure.facecolor']='#00000000'
plt.figure(figsize=(25,25))
sns.heatmap(a.corr(),annot=True,fmt='.1f')
plt.title("Corelation between the propertiesof attributes of football players")
plt.show()
```
